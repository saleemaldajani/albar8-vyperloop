### Set up the distribution as a uniform distribution from -25 to 100, starting with 10 draws

```python
%%R -o dist
min=-25
max=100
draws=10
dist=runif(draws,min,max)
hist(dist, main = "Uniform Distribution at 10 Draws")
```

### Count the proportion random draws below zero to total using which() to simulate random draws below 0

```python
%%R -i dist 

select=dist[which(dist<0)]
sim=length(select)/length(dist)
print(sim)


```

#### It looks like this is converging to the theoretical probability we calculated by hand earlier! 

### Run the same logic for a 50 draw run and observe the changes in the calculated proportion as we increase draws

```python
%%R 
min=-25
max=100
draws=50
dist=runif(draws,min,max)
hist(dist, main = "Uniform Distribution at 50 Draws")
select=dist[which(dist<0)]
sim=length(select)/length(dist)
print(sim)
```

#### Run the same logic for a 1000 draw run and... almost there, correct to the first digit! 

```python
%%R 
min=-25
max=100
draws=1000
dist=runif(draws,min,max)
hist(dist, main = "Uniform Distribution at 1000 Draws")
select=dist[which(dist<0)]
sim=length(select)/length(dist)
print(sim)
```

### As we increase the number of draws, the proportion converges to the Pr(X < 0)=25/125=1/5=0.2

```python
%%R  
min=-25
max=100
draws=1000000
dist=runif(draws,min,max)
hist(dist, main = "Uniform Distribution at 1000000 Draws")
select=dist[which(dist<0)]
sim=length(select)/length(dist)
print(sim)
```

### Set up the distribution as a normal distribution from mean=120 and variance=25, assuming high enough number of draws that the distribution will approximately converge to a gaussian ~ N(120,25)

```python
%%R 
mean=120
var=25
draws=1000000
dist=rnorm(draws,mean,var) #Set-up normal distribution and pass in parameters specified
hist(dist, main = "1000000 Draws")
select=dist[which(dist>155)] #Count number of draws above 155
sim=length(select)/length(dist) #Calculate ratio to total to estimate probability
print(sim)
```

#### To simulate random draws above 155, we count the proportion of random draws above 155 to total using which() 
##### Pr ( X > 155 ) ~ 8% !
